```r
moda = function(ese11){
    datiTab = table(ese11)
    # print(max(datiTab)) frequenza assoluta massima
    datiTab[datiTab==max(datiTab)] # ritornami tutti i dati tali che m ha frequenza massima
}
```

```r
moda(ese12)
```

```r
moda(c(1, 2, 1, 2))
```

```r
# varianza campionaria
var(ese12)
```

```r
# varianza nostra
varianza = function(array){
    media = mean(array)
    v = 0
    for (numero in array){
        v = v + (numero - media)^2
    }
    v/length(array)
}
varianza(ese12)
```

```r
sd(ese12)
```

```r
devstandard = function(array){
    sqrt(varianza(array))
}
devstandard(ese12)
```

```r
# parte 3 di 4
# caratteri bidimensionali

# tieni salvato anche il nome delle colonne
ese2 = read.csv("dati/esempio1.12.csv", sep=";", header=TRUE)
```

```r
fa = table(ese2)
fa # una cella indica ci sono CELLA casi in cui ci sono N_RIGA stanze e N_COLONNA occupanti
```

```r
margin.table(fa, 1) #frequenza marginale sul primo campo (stanze)
fac = cbind(fa, margin.table(fa, 1)) # lo aggiunge come ultima colonna (c)alla tabella 
```

```r
margin.table(fa, 2) # frequenza marginale sugli occupanti

rbind(fac, margin.table(fac, 2)) # aggiungi anche la riga di contingenza
```

```r
require("labstatR")
bubbleplot(fa)
```

```r
plot(ese2$Occupanti, ese2$Stanze) # accedo alla colonna "Nome" del CSV tramite il $
```

```r
barplot(fa, legend=TRUE, args.legend=list(title="Stanze"), xlab="Occupanti", ylab="Frequenza")
# 
```

```r
cov(ese2$Stanze, ese2$Occupanti)
```

```r
cor(ese2$Occupanti, ese2$Stanze)
```

**Covariance**	$$ \sigma _{XY}=E[(X-\mu _{X})\,(Y-\mu _{Y})]$$
**Correlation**	$$ \rho _{XY}=E[(X-\mu _{X})\,(Y-\mu _{Y})]/(\sigma _{X}\sigma _{Y})$$

```r
# parte 4 di 4
```

```r
rr = lm(ese2$Stanze ~ ese2$Occupanti)
rr
```

```r
curva = lm(ese2$Stanze ~ 1 + ese2$Occupanti + I(ese2$Occupanti^2))
curva
```

```r
plot(ese2$Stanze, ese2$Occupanti)
abline(rr, col="red")
lines(ese2$Stanze, predict(curva), col="green")
```
